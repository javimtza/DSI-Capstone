# Gathering data from BreweryDB.com

----

BreweryDB.com has quite a complete database of beers extending in over 80,000 beers worldide. In fact, they claim to be the most complete database for the beer industry. BreweryDB has a great API that is easy to work with in order to gather data, however, there are some caveats. The free API limits to a little over a thousand beers, therefore going for the subscription is the best way to go for a complete dataset. The subscription API limits you to 200 pulls per day in which you can retrieve 50 beers per pull (10,000 beers per day).

### Defining a class for pull requests
----
The following definition was retrieved from [this link](https://github.com/yarian/brewerydb) which is available as a resource on BreweryDB's [service libraries](https://www.brewerydb.com/developers/docs/library)

```python
import requests
import pandas as pd

DEFAULT_BASE_URI = "https://api.brewerydb.com/v2/"
BASE_URI = "https://api.brewerydb.com/v2/"
API_KEY = "876fdd69a9fdcc2b3a9c889b2010edd8"

simple_endpoints = ["beers", "breweries", "categories", "events",
                    "featured", "features", "fluidsizes", "glassware",
                    "locations", "guilds", "heartbeat", "ingredients",
                    "search", "search/upc", "socialsites", "styles"]

single_param_endpoints = ["beer", "brewery", "category", "event",
                          "feature", "glass", "guild", "ingredient",
                          "location", "socialsite", "style", "menu"]

class BreweryDb:

    @staticmethod
    def __make_simple_endpoint_fun(name):
        @staticmethod
        def _function(options={}):
            return BreweryDb._get("/" + name, options)
        return _function

    @staticmethod
    def __make_singlearg_endpoint_fun(name):
        @staticmethod
        def _function(id, options={}):
            return BreweryDb._get("/" + name + "/" + id, options)
        return _function

    @staticmethod
    def _get(request, options):
        options.update({"key" : BreweryDb.API_KEY})
        return requests.get(BreweryDb.BASE_URI + request, params=options).json()

    @staticmethod
    def configure(apikey, baseuri=DEFAULT_BASE_URI):
        BreweryDb.API_KEY = apikey
        BreweryDb.BASE_URI = baseuri
        for endpoint in simple_endpoints:
            fun = BreweryDb.__make_simple_endpoint_fun(endpoint)
            setattr(BreweryDb, endpoint.replace('/', '_'), fun)
        for endpoint in single_param_endpoints:
            fun = BreweryDb.__make_singlearg_endpoint_fun(endpoint)
            setattr(BreweryDb, endpoint.replace('/', '_'), fun)
```

```python
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
```

### Gathering data
----
This section was iterated multiple times. x needs to be replaced every time this is run to ensure that new pages are gathered.

```python
BreweryDb.configure(API_KEY, BASE_URI)
x = 976
first_page = BreweryDb.beers({'p':x})

data = pd.DataFrame(first_page['data'])

for i in range(x+1,x+200):  
    json = BreweryDb.beers({'p':i})
    frame = pd.DataFrame(json['data'])
    data = pd.concat([data,frame], ignore_index=True)
```

### Merging all csv files
----
Because we can only get 10,000 beers at a time we want to merge all the csv files together to create a database of all the beers we retrieved

```python
import glob
import pandas as pd
df = pd.concat([pd.read_csv(file) for file in glob.glob('BreweryDB*.csv')])   # Concatinates all csvs in our directory that start with BreweryDB
df = df.reset_index()
df = df.drop(columns=['Unnamed: 0','index'])
df = df.drop_duplicates()
df.to_csv('BreweryDB_all_beers.csv')
```

For this project, we will only need beers that actually have labels therefore creating a separate csv with only beers that have their labels with them is what we will use

```python
with_labels = df[df['labels'].isnull()==False]

with_labels.to_csv('Sarturday_Labels_16329.csv')
```
